```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
cmap=plt.get_cmap("viridis")
```

```python
import kmeans
```

```python
N = 1000
ndim = 2
```

```python
samples = np.random.randn(N, ndim)
samples[:, 0] *= 2
```

```python
k = 7
```

```python
KMC = kmeans.KMclustering(samples, k)
```

```python
plt.scatter(samples[:,0], samples[:, 1], alpha = 0.2, s=1)
for i in range(k):
    plt.scatter(
        samples[KMC.centers[i], 0], 
        samples[KMC.centers[i], 1])
```

```python
#KMC._optimize_once()
KMC.optimize()
```

```python
for i in range(k):
    p = samples[KMC.assignments==i]
    color = [cmap(float(i)/k)]
    plt.scatter(p[:, 0], p[:, 1], s=1, c=color)
    plt.scatter(
        samples[KMC.centers[i], 0], 
        samples[KMC.centers[i], 1], c=[cmap(float(i)/k)])
```
